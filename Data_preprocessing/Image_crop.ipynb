# 원본이미지 crop

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

mask=glob.glob("../../datasets/input" + '/*.png')
npy_list=[]

for i in range(len(mask)):

    img_name = mask[i][mask[i].rindex('/')+1:mask[i].rindex('.')]
    img1 = cv2.imread(mask[i])
    img1 = img_to_array(img1)
    
    #center
    img1[:,:340]=0
    img1[:,684:]=0
    
#     #lat
#     img1[:,340:684]=0
    
#     #up
#     img1[340:,:]=0
    
#     #mid
#     img1[:340,:]=0
#     img1[684:,:]=0
    
#     #blow
#     img1[:684,:]=0
    
    
    img_resize = cv2.resize(img1, (1024,1024))
        

    cv2.imwrite("../../2st/input/center/%s.png"%img_name,img_resize)
```

# 원본 mid 크롭 후 npy저장

```python
img_type="png"

for a in range(5):    
    test_label_path="../../1st/data/%d/non_aug/test_msk"%a
    result_mid_save_path="../../1st/data/%d/aug/npy/"%a
    result_mid_PIC_path="../../1st/data/%d/non_aug/test_msk/crop/mid/"%a
    print('-'*30)
    print('Creating test images...')
    print('-'*30)
    imglabels=[]
    midname=[]
    i = 0
    imgs = glob.glob(test_label_path+"/*."+img_type)

    imglabels = np.ndarray((len(imgs),1024,1024,1), dtype=np.uint8)
    
    for j, imgname in enumerate(imgs):
            midname = imgname[imgname.rindex("/")+1:]
            #print(midname)

            label = load_img(test_label_path +"/"+ midname, color_mode = "grayscale")

            label=label.resize((1024,1024))


            label = img_to_array(label)
            label[:,:340]=0
            label[:,684:]=0
            
            label=np.reshape(label,(1024,1024,1))
            imglabels[j] = label
            print(label.shape)
            plt.imshow(label,cmap='gray')
            plt.axis('off')
            plt.savefig(result_mid_PIC_path+"%d_mid.png"%j)

    print('loading done')
            
    np.save(result_mid_save_path+'test_1024_mid.npy', imglabels)
    print('Saving to imgs_test.npy files done.')
```
